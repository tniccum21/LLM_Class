```python
#LM Studio version - BASIC Chat Completion
# 
import requests
import tqdm as notebook_tqdm

# pip install python-dotenv openai
import os

url = "http://localhost:1234/v1/chat/completions"
payload = {
    "model": "openai/gpt-oss-20b",
    "messages": [ {"role":"system", "content":"You are a super helpful physics professor \
                   and youtube presenter that excels in explaining science to curious kids."},
        {"role":"user","content":"ELI10 Quantum Mechanics"}], 
    "max_tokens":"1200",
    "temperature":"0.3"
}
resp = requests.post(url, json=payload, headers={"Content-Type":"application/json"})
print(resp.json()["choices"][0]["message"]["content"])
```

```python
# HuggingFace vesrion - BASIC Chat Completion

from openai import OpenAI
import os
# pip install python-dotenv openai
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()  # reads .env in the current working directory

api_key = os.getenv("HF_TOKEN") # extract the HF_TOKEN from the .env file

if not api_key:
    raise RuntimeError("HF_TOKEN is not set. Make sure .env is in this folder or export the var.")

client = OpenAI(
    base_url="https://router.huggingface.co/v1",  # or your HF Inference Endpoint URL
    api_key=api_key,
)

resp = client.chat.completions.create(
    model="openai/gpt-oss-20b",    # the model served by that endpoint
    messages=[
        {"role":"system", "content":"You are a super helpful physics professor and youtube presenter that excels in explaining science to curious kids."},
        {"role":"user","content":"ELI10 Quantum Mechanics"}
    ],
    max_tokens=1200,
    temperature=0.3,
)
print(resp.choices[0].message.content)


```

```python
# version for GROQ
# 

from openai import OpenAI
import os
# pip install python-dotenv openai
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()  # reads .env in the current working directory

client = OpenAI(
    base_url="https://api.groq.com/openai/v1",
    api_key=os.environ["GROQ_API_KEY"],
)

resp = client.chat.completions.create(
    model="llama-3.1-8b-instant",  # Groq model name
    messages=[
        {"role":"system","content":"You are a super helpful physics professor and youtube presenter that excels in explaining science to curious kids."},
        {"role":"user","content":"ELI10 Quantum Mechanics"}
    ],
    max_tokens=1200,
    temperature=0.3,
)

print(resp.choices[0].message.content)
```

```python
#Version for Google Gemini

import os
from dotenv import load_dotenv
import google.generativeai as genai
from google.generativeai.types import GenerationConfig

load_dotenv()  # loads GOOGLE_API_KEY from .env

api_key = os.getenv("GOOGLE_API_KEY")
if not api_key:
    raise RuntimeError("GOOGLE_API_KEY is missing. Add it to your .env")

genai.configure(api_key=api_key)

# Pick a Gemini model (1.5 Pro = best quality; 1.5 Flash = faster/cheaper)
model = genai.GenerativeModel(
    model_name="gemini-2.5-flash-lite",
    system_instruction=(
        "You are a super helpful physics professor and YouTube presenter "
        "that excels in explaining science to curious kids."
    ),
)

resp = model.generate_content(
    contents=[{"role": "user", "parts": [{"text": "ELI10 Quantum Mechanics"}]}],
    generation_config=GenerationConfig(
        temperature=0.3,
        max_output_tokens=1200,
    ),
)

print(resp.text)
```
